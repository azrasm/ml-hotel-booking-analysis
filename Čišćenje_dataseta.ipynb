```python
word_to_digit = {
    "nula": "0",
    "dva": "2",
    "tri": "3",
    "četiri": "4",
    "osam": "8",
    "šest": "6"
}

import re

def decode_mixed_number(x):
    parts = re.findall(r'[0-9]+|[A-Za-zšđčćžŠĐČĆŽ]+', str(x))
    result = ""
    for p in parts:
        if p.isdigit():
            result += p
        else:
            p_lower = p.lower()
            if p_lower in word_to_digit:
                result += word_to_digit[p_lower]
            else:
                result += p
    return result


df["lead_time"] = df["lead_time"].apply(decode_mixed_number)
```

```python
month_map = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}

def convert_month(x):
    if isinstance(x, str) and x.capitalize() in month_map.values():
        return x.capitalize()

    try:
        num = int(str(x).strip())
        return month_map.get(num, x)
    except:
        return x

df["arrival_date_month"] = df["arrival_date_month"].apply(convert_month)
```

```python
import re
import operator

ops = {
    "+": operator.add,
    "-": operator.sub,
    "*": operator.mul,
    "/": operator.floordiv
}

def safe_eval(expr):
    expr = str(expr).strip()

    if expr.isdigit():
        return int(expr)

    match = re.fullmatch(r"(\d+)\s*([+\-*/])\s*(\d+)", expr)
    if not match:
        return None

    a, op, b = match.groups()
    a, b = int(a), int(b)

    return ops[op](a, b)

df["arrival_date_day_of_month"] = df["arrival_date_day_of_month"].apply(safe_eval)

df["arrival_date_day_of_month"] = df["arrival_date_day_of_month"].astype(int)
```

```python
import re

def decode_power_of_two(x):
    x = str(x).strip()

    if x.isdigit():
        return int(x)

    match = re.fullmatch(r"2\s*\^\s*(\d+)", x)
    if match:
        exp = int(match.group(1))
        return 2 ** exp

    return None

df["stays_in_week_nights"] = df["stays_in_week_nights"].apply(decode_power_of_two)
df["stays_in_week_nights"] = df["stays_in_week_nights"].astype(int)
```

```python
import re
def parse_waiting(val):
    if pd.isna(val):
        return np.nan

    if re.fullmatch(r"\d+", str(val).strip()):
        return int(val)

    m = re.match(r"(\d+)\s+mjeseci\s+i\s+(\d+)\s+dana", str(val))
    if m:
        months = int(m.group(1))
        days = int(m.group(2))
        return months * 30 + days

    return np.nan

df["days_in_waiting_list"] = df["days_in_waiting_list"].apply(parse_waiting)
df["days_in_waiting_list"] = df["days_in_waiting_list"].astype("int")
```

```python
def clean_adr(val):
    if pd.isna(val):
        return np.nan

    val = str(val).strip()
    try:
      if "+" in val:
        try:
            parts = val.split("+")
            nums = [float(p) for p in parts]
            return sum(nums)
        except:
            return np.nan

      if float(val).is_integer():
            return int(val)
      return float(val)
    except:
        return val

df["adr"] = df["adr"].apply(clean_adr)
```

```python
import pandas as pd
import re

def fix_reservation_date(val):
    if pd.isna(val):
        return val

    val = str(val).strip()

    if re.match(r"^\d{2}/\d{2}/\d{4}$", val):
        try:
            return pd.to_datetime(val, format="%d/%m/%Y").strftime("%Y-%m-%d")
        except:
            return val

    if re.match(r"^\d{4}-\d{2}-\d{2}$", val):
        return val

    return val

df["reservation_status_date"] = df["reservation_status_date"].apply(fix_reservation_date)
```
